```python
from gensim.models.keyedvectors import KeyedVectors

vectors = KeyedVectors.load_word2vec_format('../vectors/vanilla_vectors.txt', binary=False)
```

```python
vocab = [word for word in vectors.vocab]
vec = vectors[vocab]
```

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import TruncatedSVD

vec_reduced = TruncatedSVD(n_components=50, random_state=0).fit_transform(vec)
vec_embedded = TSNE(n_components=2, perplexity=40, verbose=2).fit_transform(vec_reduced)
```

```python
from sklearn.decomposition import PCA

vec_pca = PCA().fit_transform(vec)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 10))

# Plot TSNE of vocabulary with labels
plt.subplot(121)
ax = plt.axes(frameon=False)
plt.setp(ax, xticks=(), yticks=())
plt.subplots_adjust(left=0.0, bottom=0.0, right=1.0, top=0.9, wspace=0.0, hspace=0.0)
plt.scatter(vec_embedded[:, 0], vec_embedded[:, 1])

for i, txt in enumerate(vocabulary):
    ax.annotate(txt, (vec_embedded[i,0], vec_embedded[i,1]))

# Plot PCA of vocabulary with labels
subplot(122)
ax1 = axes(frameon=False)
setp(ax1, xticks=(), yticks=())
subplots_adjust(left=0.0, bottom=0.0, right=1.0, top=0.9, wspace=0.0, hspace=0.0)
plt.scatter(vec_pca[:, 0], vec_pca[:, 1])

for i, txt in enumerate(vocabulary):
    ax1.annotate(txt, (vec_pca[i,0], vec_pca[i,1]))
```
