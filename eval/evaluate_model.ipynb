```python
import csv

ratings_file_name = 'relsim/relsim_mean_ratings.csv'
comp_ratings = []
with open(ratings_file_name, newline='') as ratings:
    spamreader = csv.reader(ratings, delimiter=',')
    for row in spamreader:
        print(row)
        break
    
    for row in list(spamreader)[1:]:
        comp_ratings.append(row)
        
# print(comp_ratings[:5])
```

```python
# Evaluate the model (word2vec or GloVe) on the relational similarity ratings
# for each relational type

import numpy as np
from numpy.linalg import norm
from scipy.stats import pearsonr

num_rels = 10

# Get the ratings and predictions for each relational type

ratings_each_rel = [[] for i in range(num_rels)]
cos_each_rel = [[] for i in range(num_rels)]
euc_each_rel = [[] for i in range(num_rels)]

for row in comp_ratings:
    comp = row[2:]
    mean_rating = row[10]
    rel1 = comp[0]
    rel2 = comp[3]
    pair1 = comp[1:3]
    pair2 = comp[4:6]
#     print(pair1, pair2)
    
    if dic.has_words(*pair1) and dic.has_words(*pair2):
        reltype1 = int(rel1[:-1])
        reltype2 = int(rel2[:-1])
        if reltype1 == reltype2:   # within-type comparisons
            rel = reltype1 - 1
            cos_sim = dic.relational_sim(pair1, pair2, method='cosine')
            euc_sim = dic.relational_sim(pair1, pair2, method='euclidean')
            
            ratings_each_rel[rel].append(mean_rating)
            cos_each_rel[rel].append(cos_sim)
            euc_each_rel[rel].append(euc_sim)
#         else:
#             print('Not same type')
            
            
# Display Pearson's r between human ratings and cosine/Euclidean similarity
for reltype in range(num_rels):
    rated_sims = ratings_each_rel[reltype]
    cos_sims = cos_each_rel[reltype]
    euc_sims = euc_each_rel[reltype]
#     cos_r = pearsonr(rated_sims, cos_sims)[0]
#     euc_r = pearsonr(rated_sims, euc_sims)[0]
    
    print('Relation', reltype + 1)
    print('cosine sim =', cos_sims)
    print('Euclidean sim =', euc_sims)
    print('Sim rating = ', rated_sims)
```
